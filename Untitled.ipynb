```python
from DQN import *
from DDQN import *
from DDQN_updated import *
import gym
import pygame
env =gym.make("LunarLander-v2")
import sys
```

## 1. Deep Q Learning Agent

```python
agent_dqn = DQNAgent ( env =env, lr =5e-4,epsilon=1,epsilon_min=0.01,gamma=0.99,decay_factor=0.995,tau=1e-3)
```

```python
agent_dqn.train(2000,True)
frames= agent_dqn.play()
save_frames_as_video(frames)
```

# 2. Double Deep Q learning Agent

```python
agent_ddqn = DDQNAgent ( env =env, lr =5e-4,epsilon=1,epsilon_min=0.01,gamma=0.99,decay_factor=0.995,tau=1e-3)
agent_ddqn.train(2000,True)
frames= agent_ddqn.play()
save_frames_as_video(frames)
```

# 3. Double Deep Q learning Agent ( Updated Version )

```python
agent_ddqn2 = DDQNAgent ( env =env, lr =5e-4,epsilon=1,epsilon_min=0.01,gamma=0.99,decay_factor=0.995,tau=1e-3)
agent_ddqn2.train(2000,True)
frames= agent_ddqn2.play()
save_frames_as_video(frames)
```
